```python
##############################cellule réalisé avant le décodeur afin de déterminer la clé###################################
#"Atbash" permet de prendre un texte et d'en inverser chaque lettre, comme si il y avait un miroir au milieu de l'alphabet.
#par exemple a=z,b=y,c=x.
texte4 = """jeqeqecvnf suozvb jfk muj
dfjr fmy rvuqsk ve
itajtd mifwz nnrt
imtrvp zuh srzmzbqz tepr zn
tmsnirt imtrvp nec hw
dzpqj tjf pdecpr zl jr
ptejnt ekpb iu b
iiuyu iy ijz surg rjs ttsn
votp ac hw rzpuen jozw
rvwdvx jbo nirscyjv fi

svmkyw ve iaflss yie te
teffvv'u riznxjzvv jfk
nelrhtjrk dh sivdvjvve
yi cvb à jffrds tdp
rvwdv sebr onvnqsy zvp
zuhjwiM le wmifo wiezib nec
triot qmjvr'c onrwz
memfqg srq wdaietsq vk
"""

def position(lettre):
    pos = (alphabet.find(lettre))
    return pos

def Atbash(texte):
    texte_fin = ""
    i = 0
    while i < len(texte):
        if (position(texte[i]) != -1):
            texte_fin += alphabet[abs(position(texte[i])-25)]
        else:
            texte_fin += texte[i]
        i += 1
    return texte_fin

alphabet = "abcdefghijklmnopqrstuvwxyz"

print(Atbash(texte4))
#l'indice du texte 3 étant un peu flou, on test différente possibilité de "miroir" sur le texte 4
#on garde la fonction Atbash, pour voir si elle s'avere être utile plus tard, vu qu'elle peut traiter n'importe qu'elle texte,
#et qu'il est possible que le texte 4 soit la somme de 2 méthodes de cryptage.
```

```python
##############################cellule réalisé avant le décodeur afin de déterminer la clé###################################
#voir texte 2 pour des explications de la fonction "frequence"
texte4 ="""jeqeqecvnf suozvb jfk muj
dfjr fmy rvuqsk ve
itajtd mifwz nnrt
imtrvp zuh srzmzbqz tepr zn
tmsnirt imtrvp nec hw
dzpqj tjf pdecpr zl jr
ptejnt ekpb iu b
iiuyu iy ijz surg rjs ttsn
votp ac hw rzpuen jozw
rvwdvx jbo nirscyjv fi

svmkyw ve iaflss yie te
teffvv'u riznxjzvv jfk
nelrhtjrk dh sivdvjvve
yi cvb à jffrds tdp
rvwdv sebr onvnqsy zvp
zuhjwiM le wmifo wiezib nec
triot qmjvr'c onrwz
memfqg srq wdaietsq vk
"""

alphabet = "abcdefghijklmnopqrstuvwxyz"


def position(lettre):
    pos = (alphabet.find(lettre))
    return pos

def frequence(texte):
    res = {}
    for i in texte:
        if position(i) != -1:
            if i in res.keys():
                res[i]+=1
            else:
                res[i]=1
    for i in res:
        res[i]="{:.3f}".format(res[i]/len(texte))
    res = dict(sorted(res.items(), key=lambda caractere: caractere[1]))
    return res

print(frequence(texte4))
#on test "frequence" sur texte 4 normal, et Atbash. Après beaucoup de tentatives, on détermine qu'il ne s'agit pas de substitution
```

```python
##############################cellule réalisé avant le décodeur afin de déterminer la clé###################################
#la fonction "miroir" revoie le texte comme si il y avait un miroir à droite ou à gauche du texte, la première lettre devient la dernère, la seconde devient l'avant dernière sur la même ligne, et ainsi de suite
txt="""jeqeqecvnf suozvb jfk muj
dfjr fmy rvuqsk ve
itajtd mifwz nnrt
imtrvp zuh srzmzbqz tepr zn
tmsnirt imtrvp nec hw
dzpqj tjf pdecpr zl jr
ptejnt ekpb iu b
iiuyu iy ijz surg rjs ttsn
votp ac hw rzpuen jozw
rvwdvx jbo nirscyjv fi

svmkyw ve iaflss yie te
teffvv'u riznxjzvv jfk
nelrhtjrk dh sivdvjvve
yi cvb à jffrds tdp
rvwdv sebr onvnqsy zvp
zuhjwiM le wmifo wiezib nec
triot qmjvr'c onrwz
memfqg srq wdaietsq vk"""


def miroir(texte):
    final = ""
    ligne = ""
    for lec in range(len(texte)):
        if texte[lec] != '\n':
            ligne += texte[lec]
        else:
            final += ligne[::-1]
            final += '\n'
            ligne = ""
    final += ligne[::-1]
    return final


print(miroir(txt))
#la fonction ne donne pas de résultat intérressant en un premier temps, mais sera ensuite amélioré et réutilisé pour le déchiffrage
```

```python
##############################cellule réalisé avant le décodeur afin de déterminer la clé###################################
#voir texte 3 pour des explications de la fonction "Tronc"
def LecUpdate(IntLecteur, texte, taille):
    NewLecteur = ''
    for k in range(taille):
        NewLecteur += texte[IntLecteur + k]
    return NewLecteur

def TestMot(i, texte, lecteur, patern):
    if i == 0:
        if (lecteur == patern) and (position(texte[len(patern)+i]) == -1):
            return True
        else:
            return False

    elif i == len(texte)-len(patern):
        if (lecteur == patern) and (position(texte[i-1]) == -1):
            return True
        else:
            return False

    else:
        if (lecteur == patern) and (position(texte[len(patern)+i]) == -1) and (position(texte[i-1]) == -1):
            return True
        else:
            return False

def RepetitionTerme(texte, patern):
    res = {}
    res[patern] = [0]
    i = -1
    lecteur = LecUpdate(i, texte, len(patern))
    MePos = []
    while i != (len(texte)-len(patern)):
            i += 1
            lecteur = LecUpdate(i, texte, len(patern))
            if TestMot(i, texte, lecteur, patern) == True:
                res[patern][0] += 1
                MePos += [i]
    for j in range(len(MePos)):
        res[patern] += [MePos[j]-MePos[0]]
    res[patern][1] = MePos[0]
    return res


def position(lettre):
    pos = (alphabet.find(lettre))
    return pos

def UTM(l, m, OldListeMots):
    if position(l) != -1:
        m += l
    elif m != '':
        OldListeMots.append(m)
        m = ''
    return m, OldListeMots

def TrouverMot(texte):
    ListeMots = []
    mot = ''
    for lettre in texte:
        mot, ListeMots = UTM(lettre, mot, ListeMots)
    ListeMots.append(mot)
    return ListeMots



def Tronc(texte):
    ResFinal = {}
    for n in range(len(TrouverMot(txt))):
        if RepetitionTerme(txt, TrouverMot(txt)[n])[TrouverMot(txt)[n]][0] != 1:
            ResFinal[TrouverMot(txt)[n]] = RepetitionTerme(texte, TrouverMot(txt)[n])[TrouverMot(txt)[n]]
    ResFinal.pop('')
    return ResFinal


alphabet = "abcdefghijklmnopqrstuvwxyz"

txt= """qvjvjvxemu hflaey qup nfq
wuqi unb iefjhp ev
rgzqgw nruda mmig
rngiek afs hianayja gvki am
gnhmrig rngiek mvx sd
wakjq gqu kwvxki ao qi
kgvqmg vpky rf y
rrfbf rb rqa hfit iqh gghm
elgk zx sd iakfvm qlad
iedwec qyl mrihxbqe ur

henpbd ev rzuohh brv gv
gvuuee'f iramcqaee qup
mvoisgqip ws hreweqeev
br xey à quuiwh gwk
iedwe hvyi lmemjhb aek
afsqdrM ov dnrul drvary mvx
girlg jnqei'x lmida
nvnujt hij dwzrvghj ep
"""

print(Tronc(txt))
#à l'aide de Tronc on trouve des résultats intérressants, puis après plusieurs tentatives utilisant la même méthode que pour le texte 3 on voit que les mots s'écrivent à l'envers.
#on termine la clé puis on améliore la fonction "miroir" en "MiroirAvance" pour inverser aussi les lignes pour y donner du sens.
```

```python
###############################################cellule décodant le texte####################################################
#on import les bibliothèques, tkinter pour créer la fenêtre facilement et scrolledtext qui permet d'avoir une barre de 
#défilement (qui sera surtout utile pour le texte 4)
import tkinter as tk
from tkinter.scrolledtext import ScrolledText


#on définit les fonctions:
#"position" donne la position d'une lettre donné par rapport à ça place dans l'alphabet, si il n'est  pas trouvé
#il renvoie -1.
#cette fonction sera réutilisée pour décodé tout les textes, elle sert surtout à racourcir et rendre plus compréhensible
#les lignes
def position(lettre):
    pos = (alphabet.find(lettre))
    return pos

#les fonctions "dechiffre" suivies du numéro du texte qu'elles déchiffrent, serve à faire la partie principale de
#déchiffrement
#ici "dechiffre" déchiffre le texte 1 par chiffrement de décalage d'un pas de 1
def dechiffre1():
    resultat.delete(1.0,tk.END)
    texte = entree_texte.get()
    texte_fin = ""

    for i in range(len(texte)):
        if (position(texte[i]) != -1):
            texte_fin += alphabet[(position(texte[i])+1)%26] 
        else:
            texte_fin += texte[i]

    resultat.insert(1.0,texte_fin)


#"Mtxt" permet de directement mettre le texte cherché dans l'entré de texte, afin de faciliter les tests de décryptage
#sur le texte en question
#cette fonction sera réutilisée pour décodé tout les textes
def MTxt():
    entree_texte.delete(0,tk.END)
    entree_texte.insert(0, texte1)


#on définit les variables en dehors de fonctions:
alphabet = "abcdefghijklmnopqrstuvwxyz"

texte1 = "kd oqnbgzhm ehbghdq ztqz tm bncd ozq rtarshstshnm zkogzadshptd: bgzptd kdssqd drs qdlokzbdd ozq tmd ztsqd. tshkhrdq kz eqdptdmbd cdr kdssqdr ontq cdbncdq kd ldrrzfd."


#On utilise tkinter afin de créer la fenêtre avec 2 boutons(1 bouton pour mettre le texte dans l'entrée à décoder, 1bouton pour décoder l'entrée), 2 labels les informations sur ce que contient l'entrée et le résultat et 2 entrées, 1 entrée ou taper le texte à décoder ou où mettre le texte, et une entrée ou est mis le résultat sous forme de scrolledtext (qui permet d'avoir une barre de défilement, surtout utile pour le texte 4
racine=tk.Tk()
racine.title("Projet Crypto Texte 1")


label_texte = tk.Label(racine, text="Texte codé:")
label_texte.grid(row=0, column=0)

entree_texte = tk.Entry(racine, width=70)
entree_texte.grid(row=0, column=1)

bouton_MT1 = tk.Button(racine, text="Mettre Texte 1", command = MTxt)
bouton_MT1.grid(row=0, column=2, sticky="EW")


label_dech = tk.Label(racine, text="Texte décodé:")
label_dech.grid(row=1, column=0)

bouton_decoder = tk.Button(racine, text="Décoder texte", command = dechiffre1)
bouton_decoder.grid(row=1, column=2, sticky="EW")

resultat = ScrolledText(racine, height=5)
resultat.grid(row=1, column=1)


racine.mainloop()
```
